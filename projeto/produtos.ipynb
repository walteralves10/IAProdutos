###bibliotecas utilizadas para treinar e modificar imagem

```
from keras.models import Sequential
from keras.layers import Conv2D
from keras.layers import MaxPooling2D
from keras.layers import ZeroPadding2D
from keras.layers import Flatten
from keras.layers import Dense
from keras.layers.core import Dropout
import PIL
```

###Foram criado essas camadas sequenciais convolucionais para buscar formas/alterações  da imagem para ampliar um resultado final no calculo da acc. Explicando cada metodo abaixo:



> Dropout: Uma maneira simples de impedir que as redes neurais se sobreponham

>   ZeroPadding2D: Esta camada pode adicionar linhas e colunas de zeros no lado superior, inferior, esquerdo e direito de um tensor de imagem.

> MaxPooling2D: Operação de agrupamento máximo para dados espaciais.

> Conv2D: Essa camada cria um kernel de convolução que é convolvido com a entrada da camada para produzir um tensor de saídas

> Flatten: Aplaina a entrada. Não afeta o tamanho do lote.

```
##
classifier = Sequential()
classifier.add(Conv2D(32, (3, 3), input_shape = (64, 64, 3), activation = 'relu'))
classifier.add(ZeroPadding2D((1,1)))
classifier.add(MaxPooling2D(pool_size = (2, 2), strides=(2,2)))
classifier.add(Dropout(0.5))
```

```
##
classifier.add(ZeroPadding2D((1,1)))
classifier.add(Conv2D(32, (3, 3), activation = 'relu'))
classifier.add(ZeroPadding2D((1,1)))
classifier.add(MaxPooling2D(pool_size = (2, 2),strides=(2,2)))
#classifier.add(Flatten())
classifier.add(Dropout(0.5))
```

```
##
classifier.add(ZeroPadding2D((1,1)))
classifier.add(Conv2D(32, (3, 3), activation = 'relu'))
classifier.add(ZeroPadding2D((1,1)))
classifier.add(MaxPooling2D(pool_size = (2, 2),strides=(2,2)))
classifier.add(Dropout(0.5))
```

```
##
classifier.add(ZeroPadding2D((1,1)))
classifier.add(Conv2D(32, (3, 3), activation = 'relu'))
classifier.add(ZeroPadding2D((1,1)))
classifier.add(MaxPooling2D(pool_size = (2, 2),strides=(2,2)))
classifier.add(Dropout(0.5))
```

```
##
classifier.add(ZeroPadding2D((1,1)))
classifier.add(Conv2D(32, (3, 3), activation = 'softmax'))
classifier.add(ZeroPadding2D((1,1)))
classifier.add(MaxPooling2D(pool_size = (2, 2),strides=(2,2)))
classifier.add(Dropout(0.5))
classifier.add(Flatten())
```

###Abaixo temos 4 camadas. A primeira camada temos 128 neuronios unicos de entrada com activation relu (Unidade Linear Retificada). A segunda camada é oculta e temos 64 neuronios com a mesma activation e como camada oculta final temos 32 neuronios e a mesma activation. Por fim na camada de saida temos 4 neuronios com activation sigmoid (função com calculo para melhorar as ondas/curvas feitas nos calculos diferente da relu que calcula linearmente).
###E por fim um compilador categorical que tem especilidade com mais de 2 saidas de imagem.

```
classifier.add(Dense(units = 128, activation = 'relu'))
classifier.add(Dense(units = 64, activation = 'relu'))
classifier.add(Dense(units = 32, activation = 'relu'))
classifier.add(Dense(units = 4, activation = 'sigmoid'))
classifier.compile(optimizer = 'adam', loss = 'categorical_crossentropy', metrics = ['accuracy'])
```

###processando as imagem com keras, ampliando, colocando tamanho de escala e recortando.

```
from keras.preprocessing.image import ImageDataGenerator
train_datagen = ImageDataGenerator(rescale = 1./255,
 shear_range = 0.2,
 zoom_range = 0.2,
 horizontal_flip = True)
```

###Buscando arquivos do dataset no google drive e passando e recortando as imagens para as variaveis atribuidas.

```
test_datagen = ImageDataGenerator(rescale = 1./255)
training_set = train_datagen.flow_from_directory('./gdrive/My Drive/RedesNeurais/produtos/produtos/train',
 target_size = (64, 64),
 batch_size = 32,
 class_mode = 'categorical')
test_set = test_datagen.flow_from_directory('./gdrive/My Drive/RedesNeurais/produtos/produtos/test',
 target_size = (64, 64),
 batch_size = 32,
 class_mode = 'categorical')
```

###Treinando atraves das imagens com 8 epocas percorrendo 200 vezes treinando e depois salvando em um arquivo ".H5"

```
classifier.fit_generator(training_set,
 steps_per_epoch = 200,
 epochs = 6,
 validation_data = test_set,
 validation_steps = 32)
```
